```sage
def CreateBlockMatrix(dimensions, blockshape, blocklist):
    '''Create one matrix out of several specified blocks'''
    return matrix.block(dimensions[0],dimensions[1], [blocklist[t] for t in blockshape])

def is_zeroone(matrix):
    '''Check if a matrix is a zero-one matrix'''
    for row in matrix:
        for entry in row:
            if entry not in {0,1}:
                return False

    return True

I = identity_matrix(2)
O = zero_matrix(2)
N = matrix([[0,0],[1,1]])
J = ones_matrix(2)
M = N.transpose()
diagblocklist = {'I':I, 'O':O, 'N':N, 'M':M}
fullblocklist = list(diagblocklist.values()) + [J-X for X in diagblocklist.values()]
Y = J - 2*I

orthog = CreateBlockMatrix((2,1), [0,1], [J, Y])
perdiag = dict()

for diag in diagblocklist.keys():
    fourblocks = []
    for lower in diagblocklist.keys():
        for upper in diagblocklist.keys():
            fourblock = CreateBlockMatrix((2,2), [0,1,2,0], [diagblocklist[diag], diagblocklist[upper], diagblocklist[lower]])
            X = orthog.transpose()*fourblock*orthog/4
            if is_zeroone(X):
                if diag in ['N', 'M']:
                    fourblocks.append([upper, lower,X])

    perdiag[diag] = fourblocks
    print(perdiag[diag])

allfourblocks = []
for diag in fullblocklist:
    for lower in fullblocklist:
        for upper in fullblocklist:
            fourblock = CreateBlockMatrix((2,2), [0,1,2,0], [diag, upper, lower])
            X = orthog.transpose()*fourblock*orthog/4
            if is_zeroone(X):
                allfourblocks.append([upper, lower])
                
print(M)
print(len(allfourblocks))
```
